```python
import random
```

```python
# defining the fitness function:
def fitness_function(x,y):
    return (4*(x**2))-(y**3)+700
```

```python
# Generating an initial population:
def init_pop(pop_size):
    pop=[]
    for _ in range(pop_size):
        x=random.uniform(-10,10)
        y=random.uniform(-10,10)
        pop.append((x,y))
    return pop
```

```python
# Select parents for reproduction using tournament selection:
def select_parents(pop,num_par):
    selected_parents=[]
    for _ in range(num_par):
        tourn_size=5
        tourn=random.sample(pop,tourn_size)
        best_ind=max(tourn,key=lambda x:fitness_function(*x))
        selected_parents.append(best_ind)
    return selected_parents
```

```python
# Perform crossover to create parents:
def crossover(par,num_ofs):
    offsprings=[]
    for _ in range(num_ofs):
        p1,p2=random.sample(par,2)
        x_ofs=(p1[0]+p2[0])/2
        y_ofs=(p1[1]+p2[1])/2
        offsprings.append((x_ofs,y_ofs))
    return offsprings
```

```python
# Perform mutatation to introduce genetic diversity:
def mutation(ind,m_rate):
    x,y=ind
    if random.random()<m_rate:
        x+=random.uniform(-0.5,0.5)
#         print(f"x mutated for {ind}")
    if random.random()<m_rate:
        y+=random.uniform(-0.5,0.5)
#         print(f"y mutated for {ind}")
    return (x,y)
```

```python
# DRIVER FUNCTION:

#setting up genetic algorithm parameters:
m_rate=0.1
pop_size=100
num_gen=100
num_par=50
num_ofs=pop_size-num_par
```

```python
# Setting up initial population:
population=init_pop(pop_size)
population[:5]
```

```python
x_axis=[]
y_axis=[]
# Generations loop:
for gen in range(num_gen):
    # evaluating the fitness of each function:
    f_Score=[fitness_function(*i) for i in population]
    
    # select parents:
    parents=select_parents(population,num_par)
    
    # performing crossover to create offsprings:
    offsprings=crossover(parents,num_ofs)
    
    # mutate the offsprings:
    offsprings=[mutation(ind,m_rate) for ind in offsprings]
    
    # replacing the old population:
    population=parents+offsprings
    
    # find and print the best solution:
    population.sort(key=lambda x:fitness_function(*x))
    population.reverse()
    best=population[0]
    print(f"gen{gen}: {best} {fitness_function(*best)}")
    x_axis.append(gen)
    y_axis.append(fitness_function(*best))
```
