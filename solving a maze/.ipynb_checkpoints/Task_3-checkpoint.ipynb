```python
def sol_zone(sol,score):
    x=6
    y=18
    p,q=sol[-1]
    if score==(5**0.5):
        if ((x,y+1) in sol) or ((x,y-1) in sol) or ((x+1,y) in sol) or ((x-1,y) in sol) or ((x+1,y+1) in sol) or ((x-1,y-1) in sol) or ((x+1,y-1) in sol) or ((x-1,y+1) in sol):
            score=1.75
        else:
            score=2.5
    elif score==2:
        if ((x,y+1) in sol) or ((x,y-1) in sol) or ((x+1,y) in sol) or ((x-1,y) in sol) or ((x+1,y+1) in sol) or ((x-1,y-1) in sol) or ((x+1,y-1) in sol) or ((x-1,y+1) in sol):
            score=2
        else: 
            score=2.5
    elif score==(2**0.5):
        if ((x,y+1) in sol) or ((x,y-1) in sol) or ((x+1,y) in sol) or ((x-1,y) in sol) or ((x+1,y+1) in sol) or ((x-1,y-1) in sol) or ((x+1,y-1) in sol) or ((x-1,y+1) in sol):
            possibilities=maze[(p,q)]
            score=2.2
            for new in possibilities:
                if (x,y) in maze[new]:
                    score=1.5
        else:
            score=2.2
    elif score==1:
        if (x,y) in maze[(p,q)]:
            score=1
        else:
            score=2.1
    else:
        pass
    return score
```

```python
def fitness_score(sol):
    x,y=sol[-1]
    x=abs(x-6)**2
    y=abs(y-18)**2
    score=(x+y)**0.5
    if score in [5**0.5,2,2**0.5,1]:
        score=sol_zone(sol,score)
    return score
```

```python
def create_pop(pop_size):
    sols=[]
    count=0
    while count<pop_size:
        sol=[(0,0)]
        pt=(0,0)
        while True:
            pt_new=random.choice(maze[pt])
            if pt_new not in sol:
                sol.append(pt_new)
                pt=pt_new
                continue
            else:
                break
        if sol not in sols:
            sols.append(sol)
            count+=1
    return sols
```

```python
def crossover(p1,p2):
    split_pt=0
    x=None
    for i in range(len(p1)):
        x=p1[i]
        if x in p2:
            split_pt=i
    x=p1[split_pt]
    j=0        
    for j in range(len(p2)):
        y=p2[j]
        if y==x:
            break

    offspring=p1[:split_pt].copy()
    for y in p2[j:]:
        if y not in offspring:
            offspring.append(y)
        else:
            break
    return offspring
```

```python
def mutation(offspring,m_rate):
    i=random.random()
    if i<m_rate:
        x=offspring[-1]
        while True:
            new_node=random.choice(maze[x])
            if new_node not in offspring:
                offspring.append(new_node)
                break
            if (len(maze[x])==2 and ((maze[x][0] in offspring) and (maze[x][1] in offspring))):
                break
    return offspring
```

```python
pop_size=30
m_rate=0.93
sols=create_pop(pop_size)
for sol in sols:
    print(sol)
```

```python
count=0
while True:
    sols.sort(key=lambda x:fitness_score(x))
    best_sol=sols[0]
    if fitness_score(best_sol)==0:
        break
    count+=1
    print(f"{best_sol},{fitness_score(best_sol)}\n")
    par_len=len(sols)//2
    ofs_len=30-par_len
    parents=sols[:par_len].copy()
    offsprings=[]
    for _ in range(ofs_len):
        p1,p2=random.sample(parents,2)
        offspring=crossover(p1,p2)
        offspring=mutation(offspring,m_rate)
        offsprings.append(offspring)
        p1,p2=random.sample(parents,2)
        offspring=crossover(p1,p2)
        offspring=mutation(offspring,m_rate)
        offsprings.append(offspring)
    sols=parents+offsprings
    
x=sols[0]
print(fitness_score(x))
print(x)
print(f'solution obtained in {count} iterations')
```
