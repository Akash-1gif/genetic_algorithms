```python
def fitness_score(sol):
    x,y=sol[-1]
    x=abs(x-6)**2
    y=abs(y-18)**2
    score=(x+y)**0.5
    return score
```

```python
def create_pop():
    sols=[]
    count=0
    while count<30:
        sol=[(0,0)]
        pt=(0,0)
        while True:
            pt_new=random.choice(maze[pt])
            if pt_new not in sol:
                sol.append(pt_new)
                pt=pt_new
                continue
            else:
                break
        if sol not in sols:
            sols.append(sol)
            count+=1
    return sols
```

```python
def crossover(p1,p2):
    split_pt=0
    x=None
    for i in range(len(p1)):
        x=p1[i]
        if x in p2:
            split_pt=i
    x=p1[split_pt]
    print(x,p1[split_pt])
    j=0        
    for j in range(len(p2)):
        y=p2[j]
        if y==x:
            break
#     print(p1[:split_pt])
#     print(p2[j:])
#     offspring=p1[:split_pt]+p2[j:]
    offspring=p1[:split_pt].copy()
    for y in p2[j:]:
        if y not in offspring:
            offspring.append(y)
        else:
            break
    return offspring
```

```python
def mutation(offspring,m_rate):
    i=random.random()
    print(i)
    if i>m_rate:
        offspring.pop()
        x=offspring[-1]
        diverse_gene=[]
        count=0
        for t in maze[x]:
            if t in offspring:
                count+=1
            else:
                diverse_gene.append(t)
        if count<len(maze[x]):
            offspring.append(random.choice(diverse_gene))
            
    return offspring
```

```python
sols=create_pop()
for sol in sols:
    print(sol)
```

```python
sols.sort(key=lambda x: fitness_score(x))
for sol in sols:
    print(sol)
```
