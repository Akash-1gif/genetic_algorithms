```python
import random
```

```python
object_weights=[6,18,12,14,19,7,5]
object_profits=[100,300,250,140,250,100,120]
threshold_weight=50
```

```python
# Taking initial population as 10:
pop=[]
iter_count=0
while len(pop)<10:
    iter_count+=1
    print(f'iteration count={iter_count}')
    consider=[0,1]
    chromosome=[]
    sums=0
    profits=0
    for i,j in zip(object_weights,object_profits):
        x=random.choice(consider)
        chromosome.append(x)
        if x==1:
            sums+=i
            profits+=j
    if sums<=threshold_weight:
        pop.append((chromosome,sums,profits))
```

```python
pop.sort(key=lambda x:x[2])
pop.reverse()
pop_size=10
pop
```

```python
for _ in range(2000):
    pop.sort(key=lambda x:x[2])
    pop.reverse()
    new_pop_len=int(0.4*pop_size)
    new_pop=pop[:new_pop_len].copy()
    parents=new_pop.copy()
    competitors=pop[new_pop_len:].copy()
    pairs=[]
    while len(parents)>0:
        x1=parents.pop(len(parents)-1)
        x2=parents.pop(len(parents)-1)
        pairs.append([x1,x2])

    while True:
        s1=0
        w1=0
        s2=0
        w2=0
        for x in pairs:
            x1=x[0][0]
            x2=x[1][0]
            chromosome1=[]
            chromosome2=[]
            for i in range(len(x1)):
                d=random.random()
                if d<=0.45:
                    chromosome1.append(x1[i])
                    chromosome2.append(x2[i])
                elif d<=0.9:
                    chromosome1.append(x2[i])
                    chromosome2.append(x1[i])
                else:
                    chromosome1.append(random.choice([0,1]))
                    chromosome2.append(random.choice([0,1]))
    
        for i in range(len(chromosome1)):  
            if chromosome1[i]==1:
                s1+=object_profits[i]
                w1+=object_weights[i]
        
        for i in range(len(chromosome2)):  
            if chromosome2[i]==1:
                s2+=object_profits[i]
                w2+=object_weights[i]
            
        if w1<=threshold_weight and w2<=threshold_weight:
            competitors.append((chromosome1,w1,s1))
            competitors.append((chromosome2,w2,s2))
            break
        else:
            continue
    
    competitors.sort(key=lambda x:x[2])
    competitors.reverse()
    for x in range(int(0.6*pop_size)):
        new_pop.append(competitors[x])
    
    new_pop.sort(key=lambda x:x[2])
    new_pop.reverse()
    pop=new_pop
    print(new_pop[0])
```
